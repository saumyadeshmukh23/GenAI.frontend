```python
import pandas as pd

# Load the dataset to inspect the contents
file_path = '/content/drive/MyDrive/GenAI_Parser_Data/Base_Dataset_for_parser_GenAI.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows to understand its structure
df.head()
```

```python
import random
from datetime import datetime, timedelta

# Generate random dates for the DOB column
start_date = datetime(1950, 1, 1)
end_date = datetime(2022, 12, 31)

# Function to generate a random date between two dates
def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end - start).days))

# Add random dates to the 'DOB' column for non-NaN entries
df['DOB'] = df['DOB'].apply(lambda x: random_date(start_date, end_date) if pd.isna(x) else x)

df.head()
```

```python
# Extract symptom data and their intensities
symptoms = {
    'S1': {'Fever': set(df.iloc[1, 2:5].dropna().tolist())},
    'S2': {'Cough': set(df.iloc[1, 6:8].dropna().tolist())},
    'S3': {'Cold': set(df.iloc[1, 9:11].dropna().tolist())}
}

# Convert sets to strings for better readability in the output
symptoms = {
    'S1': {f'Fever:{intensity}' for intensity in ['mild', 'low', 'High']},
    'S2': {f'Cough:{intensity}' for intensity in ['mild', 'low', 'High']},
    'S3': {f'Cold:{intensity}' for intensity in ['mild', 'low', 'High']}
}

symptoms
```
